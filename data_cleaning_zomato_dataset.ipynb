```python
final_df = df.merge(country_df, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
cuisine_column = final_df['Cuisines']
unique_cuisines = set()

# Split cuisines where there are multiple per restaurant and add to set
for cuisines in cuisine_column.dropna():
    for cuisine in cuisines.split(', '):
        unique_cuisines.add(cuisine)

# Convert set to sorted list
sorted_unique_cuisines = sorted(unique_cuisines)

# Display the result
sorted_unique_cuisines
```
